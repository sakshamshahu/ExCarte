```python
idsToKeep = []

cnt = 0
cntTrue = 0
for row in data.iter_rows(named=True):
    # print(row['id'], row['address'])
    # print(row)
    x = row['address'].split(',')
    if len(x) >= 4 and x[-4]:
        chunks = x[-4].lower().split(' ')
        if 'jayanagar' in chunks or 'koramangala' in chunks or 'jayanagar' in chunks or 'hsr' in chunks :
            print(chunks)
            cntTrue += 1
            idsToKeep.append(row['id'])
    else:  
        continue

print(cntTrue)

```

```python
filtered_data = data.filter(pl.col('id').is_in(idsToKeep))
```
